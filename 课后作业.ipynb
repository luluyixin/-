将MA的ndarray数据转换成Series格式，并将价格和MA值用Matplotlib展示出来

```python
s = pd.Series(MA,index=data['600036.XSHG'].index)#定义索引
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 7))
plt.plot(data['600036.XSHG'].close)#导入数据
plt.plot(s)
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

用talib计算50只股票的周期为5的ROCR100，生成Dataframe，并将前5只股票的 ROCR100（参数timeperiod=20）用一张图显示出来。

```python
import talib.abstract as ta
import talib as tb
plt.figure(figsize=(15, 7))
date_index = pd.read_excel( 'sz50.xlsx' ,index_col = 'datetime').index#定义索引
columns_index = pd.read_excel('sz50.xlsx').columns
_dict = {}
for key,value in data.items():
    try:
        index = value.index#
        _dict[key] = pd.Series(tb.ROCR100(value.close.values, timeperiod = 20), index = index)
    except AttributeError:
        pass
df = pd.DataFrame(_dict)#生成Dataframe
list = df.columns.values.tolist()
for i in range(5):
    plt.plot(df[list[i]],label = list[i])
plt.legend()
plt.show()
```
